```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
```

Crea la grilla con celulas vivas aleatoriamente

```python
def create_grid(N):
  return np.random.choice([0, 1], size=(N, N))
```

```python
def update_grid(grid):
    N = grid.shape[0]
    new_grid = grid.copy()

    for i in range(N):
        for j in range(N):
            # Contar las c√©lulas vecinas vivas
            total = int((grid[i, (j-1)%N] + grid[i, (j+1)%N] + grid[(i-1)%N, j] + grid[(i+1)%N, j] +
                         grid[(i-1)%N, (j-1)%N] + grid[(i-1)%N, (j+1)%N] + grid[(i+1)%N, (j-1)%N] + grid[(i+1)%N, (j+1)%N]))

            # Aplicar las reglas del Juego de la Vida
            if grid[i, j] == 1:
                if total < 2 or total > 3:
                    new_grid[i, j] = 0
            else:
                if total == 3:
                    new_grid[i, j] = 1

    return new_grid
```

Visualizacion

```python
def animate_game_of_life(grid, frames=100, interval=200):
    #Visualizacion
    fig, ax = plt.subplots()
    img = ax.imshow(grid, cmap='binary')

    def update(frame):
        nonlocal grid
        grid = update_grid(grid)
        img.set_data(grid)
        return [img]

    ani = animation.FuncAnimation(fig, update, frames=frames, interval=interval, blit=True)
    return ani

#Ejecutamos
N = 50
grid = create_grid(N)
ani =animate_game_of_life(grid)
HTML(ani.to_jshtml())
```
